<h3><strong>Imports</strong></h3>

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

<h3><strong>HTTP Request</strong></h3>

<h3>store website is variable</h3>

```python
website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops?fts=laptops'
```

<h3>Get Request</h3>

```python
response = requests.get(website)
```

<h3>Status Code</h3>

```python
response.status_code
```

<h3>Soup Object</h3>

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

<h3>Results</h3>

```python
results = soup.find_all('div', {'class':'OfferBox'})
```

```python
len(results)
```

```python
results [0]
```

<h3>Target necessary data</h3>

```python
# Name
# Price
# Review Star Rating
# product link
# product Details
```

<h3>Name</h3>

```python
results[0].find('a', {'class': 'offerboxtitle'}).get_text()
```

<h3>Price</h3>

```python
results[0].find('span', {'class': 'offerprice'}).get_text()
```

<h3>Review Star Rating</h3>

```python
results[0].find('span', {'class': 'stars-container stars-100'}).get_text()
```

<h3>Relative URL</h3>

```python
# extract the relative URL
relative_url = results[0].find('a', {'class':'offerboxtitle'}).get('href')
```

```python
# provide the root element inside a variable
root_url = 'https://www.laptopsdirect.co.uk'
```

```python
#combine root element and relative URL
url_combine = root_url + relative_url 
```

```python
url_combine
```

<h3>Product Details</h3>

```python
results[0].find('div', {'class':'productInfo'}).get_text().strip().replace('\n', ', ')
```

<h3>Put everything together inside a for-loop</h3>

```python
product_name = []
product_price = []
product_rating = []
relative_url = []
product_details = []


for result in results:
    
    try:
        product_name.append(result.find('a', {'class': 'offerboxtitle'}).get_text())
    except:
        product_name.append('n/a')
      
    #price
    try:
        product_price.append(result.find('span', {'class': 'offerprice'}).get_text())
    except:
        product_price.append('n/a')
     
    #review star rating
    try:
        product_rating.append(result.find('span', {'class': 'stars-container stars-100'}).get_text())
    except:
        product_rating.append('n/a')
        
    # relative URL
    try:
        relative_url.append(result.find('a', {'class':'offerboxtitle'}).get('href'))
    except:
        relative_url.append('n/a')
        
    # product details
    try:
        product_details.append(result.find('div', {'class':'productInfo'}).get_text().strip().replace('\n', ', '))
    except:
        product_details.append('n/a')
```

<h3>Combine URLs</h3>

```python
url_combined = []

for link in relative_url:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

```python
#url_combined
```

<h3>Create Pandas Dataframe</h3>

```python
product_overview = pd.DataFrame({'name': product_name, 'price': product_price, 'Rating': product_rating, 
                                'link': url_combined, 'Details': product_details})
```

```python
product_overview
```

<h3>Output in Excel</h3>

```python
product_overview.to_excel('results_single_page.xlsx', index=False)
```

<h3>Part 2 - Pagination - Scrape 20 Pages</h3>

```python
product_name = []
product_price = []
product_rating = []
relative_url = []
product_details = []

for i in range(1,21):
    #website
    website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops?pageNumber=' + str(i)
   # request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('div', {'class':'OfferBox'})

    # Loop through results
    for result in results:
    
        try:
            product_name.append(result.find('a', {'class': 'offerboxtitle'}).get_text())
        except:
            product_name.append('n/a')
      
        #price
        try:
            product_price.append(result.find('span', {'class': 'offerprice'}).get_text())
        except:
            product_price.append('n/a')
     
        #review star rating
        try:
            product_rating.append(result.find('span', {'class': 'stars-container stars-100'}).get_text())
        except:
            product_rating.append('n/a')
        
        # relative URL
        try:
            relative_url.append(result.find('a', {'class':'offerboxtitle'}).get('href'))
        except: 
            relative_url.append('n/a')
        
        # product details
        try:
            product_details.append(result.find('div', {'class':'productInfo'}).get_text().strip().replace('\n', ', '))
        except:
            product_details.append('n/a')

```

```python
url_combined = []

for link in relative_url:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

```python
product_overview = pd.DataFrame({'name': product_name, 'price': product_price, 'Rating': product_rating, 
                                'link': url_combined, 'Details': product_details})
```
